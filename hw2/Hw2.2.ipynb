```python
df.drop(['Спеціалізація','Університет','Предметна область','Валюта','cls'], axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.shape
```

```python
python_data = df[df['Мова програмування'] == 'Python']
python_data
```

```python
python_data.shape
```

```python
new_df = python_data.groupby('Посада')['Зарплата на місяць'].agg(['min', 'max'])
new_df
```

```python
def fill_avg_salary(row):
    return (row['min'] + row['max']) / 2

new_df['avg'] = new_df.apply(fill_avg_salary, axis=1)
new_df['avg'].describe()
```

```python
new_df.to_csv('res.csv')
```
